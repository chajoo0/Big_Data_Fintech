```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#(1) 한글 기본 설정(이 스크립트 전체가 영향을 받음)
mpl.rcParams["font.family"] = 'NanumGothic'

#(2) 지정한 곳에서만 한글 설정을 유효하게
font_path = 'C:/Windows/Fonts/나눔고딕L.ttf'
font_prop = fm.FontProperties(fname= font_path)
font_prop.set_style('normal')
font_prop.set_size('20')
plt.figure(1)
plt.plot([1,2,3], [2,5,9])

#X축 레이블은 설정(1) 적용
plt.xlabel('한글 X축 레이블')

#Y축은 레이블은 설정(2) 적용
plt.ylabel('한글 Y축 레이블', fontproperties = font_prop)
plt.show()
```

```python
x = np.arange(-np.pi, np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x) 
plt.plot(x, y1, label='sin')
plt.plot(x, y2, 'r*', markersize=1, label='cosine')

#각 축의 레이블을 그림
plt.xlabel('X-axis'); 
plt.ylabel('Y-axis')
#범례를 그림
plt.legend(loc='best')
#그리기 실행
plt.show()
```

### 등고선

```python
from matplotlib import cm
from matplotlib.mlab import bivariate_normal
#2차원 메시를 작성
N=200
x = np.linspace(-3.0, 3.0, N)
y = np.linspace(-2.0, 2.0, N)
X, Y = np.meshgrid(x,y)

#2변량 정규 분포로 2차원 분포 데이터를 생성
z = 15* (bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)-bivariate_normal(X, Y, 1.5, 0.5, 1, 1))
```

```python
X,Y
```

```python
#z의 값을 농담값으로 삼아 이미지를 나타냄.
im = plt.imshow(z, interpolation = 'bilinear', origin = 'lower', extent = (-3, 3, -2, 2))
plt.show()
```

```python
im = plt.imshow(z, interpolation = 'bilinear', origin = 'lower', extent = (-3, 3, -2, 2))

#등고선 표시
levels = np.arange(-3, 2.5, 0.5)
ctr = plt.contour(z, levels, colors = 'k', origin = 'lower', linewidths = 2, extent=(-3, 3, -2, 2))

#등고선에 레이블 인라인 표시
plt.gca().clabel(ctr, levels, inline=1, colors='black',fmt='%1.1f', fontsize=14)

plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

#데이터를 3차원 플롯으로 플로팅
fig = plt.figure(1) 

ax =fig.add_subplot(111, projection ='3d')   

surf = ax.plot_surface(X,Y,z, cmap=cm.Greens)
ax.set_zlim3d(-3.01, 3.01)
plt.colorbar(surf, shrink = 0.5, aspect=10)
plt.show()
```
