### 등고선

```python
from matplotlib import cm
from matplotlib.mlab import bivariate_normal
#2차원 메시를 작성
N=200
x = np.linspace(-3.0, 3.0, N)
y = np.linspace(-2.0, 2.0, N)
X, Y = np.meshgrid(x,y)

#2변량 정규 분포로 2차원 분포 데이터를 생성
z = 15* (bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)-bivariate_normal(X, Y, 1.5, 0.5, 1, 1))
```

```python
X,Y
```

```python
#z의 값을 농담값으로 삼아 이미지를 나타냄.
im = plt.imshow(z, interpolation = 'bilinear', origin = 'lower', extent = (-3, 3, -2, 2))
plt.show()
```

```python
im = plt.imshow(z, interpolation = 'bilinear', origin = 'lower', extent = (-3, 3, -2, 2))

#등고선 표시
levels = np.arange(-3, 2.5, 0.5)
ctr = plt.contour(z, levels, colors = 'k', origin = 'lower', linewidths = 2, extent=(-3, 3, -2, 2))

#등고선에 레이블 인라인 표시
plt.gca().clabel(ctr, levels, inline=1, colors='black',fmt='%1.1f', fontsize=14)

plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

#데이터를 3차원 플롯으로 플로팅
fig = plt.figure(1) 

ax =fig.add_subplot(111, projection ='3d')   

surf = ax.plot_surface(X,Y,z, cmap=cm.Greens)
ax.set_zlim3d(-3.01, 3.01)
plt.colorbar(surf, shrink = 0.5, aspect=10)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# conda install -c conda-forge proj4
# conda install -c conda-forge basemap 으로 설치
# conda install -c conda-forge basemap-data-hires
from mpl_toolkits.basemap import Basemap
data = pd.read_csv('Haiti.csv')
```

```python
print(data.describe()) # describe 메서드를 호출하여 이상한 위치가 있는지 확인
```

```python
data = data[(data.LATITUDE > 18) & (data.LATITUDE < 20) 
            & (data.LONGITUDE > -75) & (data.LONGITUDE < -70)
            & data.CATEGORY.notnull()]
```

```python
def to_cat_list(catstr):
    stripped = (x.strip() for x in catstr.split(','))
    return [x for x in stripped if x]

def get_all_categories(cat_series):
    cat_sets = (set(to_cat_list(x)) for x in cat_series)
    return sorted(set.union(*cat_sets))

def get_english(cat):
    code, names = cat.split('.')
    if '|' in names:
        names = names.split('|')[1]
    return code, names.strip()
```

```python
# test if it works
print(get_english('2. Urgences logistiques | Vital Lines'))
```

```python
all_cats = get_all_categories(data.CATEGORY)
#제네레이터 표현
english_mapping = dict(get_english(x) for x in all_cats)
print(english_mapping['2a'])
print(english_mapping['6c'])
```

```python
def get_code(seq):
    return [x.split('.')[0] for x in seq if x]
```

```python
import numpy as np
all_codes = get_code(all_cats)   # all_codes에 앞에 코드들만 저장
code_index = pd.Index(np.unique(all_codes))   # code에 인덱스붙이기 
# 데이터와 같은 index 가지는 np zero mtx 만들기
dummy_frame = pd.DataFrame(np.zeros((len(data), len(code_index))),   
                            index=data.index, columns=code_index)
```

```python
# row와 코드 이름에 대응되는 dummy_frame의 값은 1로 저장, 원래 데이터와 조인
for row, cat in zip(data.index, data.CATEGORY):
    codes = get_code(to_cat_list(cat))
    dummy_frame.loc[row, codes] = 1
    
data = data.join(dummy_frame.add_prefix('category_'))
```

```python
data.iloc[:, 10:15]
```

```python
def basic_haiti_map(ax=None, lllat=17.25, urlat=20.25, lllon=-75, urlon=-71):
    #create polar stereographic Basemap instance
    m = Basemap(ax=ax, projection='stere',
                lon_0=(urlon+lllon)/2,
                lat_0=(urlat+lllat)/2,
                llcrnrlat=lllat, urcrnrlat=urlat,
                llcrnrlon=lllon, urcrnrlon=urlon,
                resolution='f')
    #draw coastlines, state and country boundaries, edge of map
    m.drawcoastlines()
    m.drawstates()
    m.drawcountries()
    return m 
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,10))
fig.subplots_adjust(hspace=0.05, wspace=0.05)
to_plot = ['2a', '1', '3c', '7a']
lllat = 17.25; urlat = 20.25; lllon = -75; urlon= -71
for code, ax in zip(to_plot, axes.flat):
    m = basic_haiti_map(ax, lllat=lllat, urlat=urlat, lllon=lllon, urlon=urlon)
    cat_data = data[data['category_%s' % code] == 1]
    
    #compute map proj coordinates
    x, y = m(np.array(cat_data['LONGITUDE']), np.array(cat_data['LATITUDE']))
    m.plot(x, y, 'k.', alpha =  0.5)
    ax.set_title('%s: %s' % (code, english_mapping[code]))
plt.show()
```
